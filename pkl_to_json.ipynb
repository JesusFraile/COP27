```python
import pandas as pd
```

```python
path='/data/jfraile/Programs/COP27/final_dataset/no_text_objectives_cop27_full_es_20.pkl'
df=pd.read_pickle(path)
df['objectives'] = df['objectives'].apply(lambda x: x[0] if isinstance(x, list) and len(x) == 1 and isinstance(x[0], list) else x)
```

```python
df
```

```python
dic={
1: "Reinforce one's own image, presence, ideology, or values.",
2: "Legitimise decisions.",
3: "Seek unity against a hostile enemy.",
4: "Weaken the opponent's image, presence, identity, ideology, or values.",
5: "Delegitimise the opponent's decisions or policies.",
6: "Promote social change or activate new policies.",
7: "Stop social change or defend the status quo."
}
```

```python
def change_number_to_text(array):
    return [dic[x] for x in array if 1 <= x <= 7]
```

```python
df['objectives_text']=df['objectives'].apply(change_number_to_text)
```

```python
df[['id', 'objectives_text']].to_json('/data/jfraile/Programs/COP27/final_dataset/objectives_cop27_full_es_20.json', orient='records', indent=2, force_ascii=False)
```
