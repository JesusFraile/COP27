```python
obj_embeddings = torch.load("embeddings/obj_embeddings_20000.pt")
dic_embeddings = torch.load("embeddings/dic_embeddings_20000.pt")
```

```python
# Compute all pairwise cosine similarities
similarity_matrix = cosine_similarity(obj_embeddings, dic_embeddings)

# Create DataFrame with indices for dic_values
df_sim = pd.DataFrame(
    similarity_matrix,
    columns=[i+1 for i in range(len(dic_values))]
)

# Add obj_list as a column (not as index)
df_sim.insert(0, "sentence", obj_list)
```

```python
threshold=0.8
matches={}
cols=list(range(1,21))
for idx, row in df_sim.iterrows():
    # print('='*20)
    for c in cols:
        if row[c]>=threshold:
            print(f'{idx} -> {row['sentence']} -> {dic[c]} -> {row[c]}')
            matches[row["sentence"]] = c
```

```python
# Update dataframe using the matches dictionary
for i in range(len(df)):
    narratives = df.at[i, 'narratives']
    for j in range(len(narratives)):
        sentence = narratives[j]
        if sentence in matches:
            narratives[j] = matches[sentence]  # Replace sentence with corresponding code
    df.at[i, 'narratives'] = narratives
```

```python
import numpy as np

threshold = 0.8
n = len(df_self_sim)

# Extract similarity values (excluding the 'sentence' column)
sim_matrix = df_self_sim.drop(columns=["sentence"]).to_numpy()

# Build adjacency matrix
adj_matrix = (sim_matrix > threshold).astype(int)
np.fill_diagonal(adj_matrix, 0)

visited = set()
groups = {}  # representative -> set of related indices

for i in range(n):
    if i not in visited:
        group = set(np.where(adj_matrix[i] == 1)[0])
        group.add(i)
        visited.update(group)
        groups[i] = group

# Display summary
print("Representatives and group sizes:\n")
for rep, members in groups.items():
    print(f"Representative {rep} ({df_self_sim.loc[rep, 'sentence']!r}) ‚Üí {len(members)} sentences")
```

```python
# Sort representatives by group size (descending)
sorted_groups = sorted(groups.items(), key=lambda x: len(x[1]), reverse=True)

print("Representatives sorted by group size:\n")
for rep, members in sorted_groups:
    print(f"Representative {rep} ({df_self_sim.loc[rep, 'sentence']!r}) ‚Üí {len(members)} sentences")
```
