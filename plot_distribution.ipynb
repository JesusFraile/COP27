```python
import pandas as pd
```

```python
df=pd.read_pickle('/data/jfraile/Programs/COP27/datasets_with_answer/cop_no_guide_Tweets_per user_id-first_peak_7_11_to_12_11_2022.csv.pkl')
len_dataset=len(df)
df['Count'] = df['Count'].str.replace(',', '').astype(float).astype(int)
total_interaction=df['Count'].sum()
narratives_list=[i for row in df['narratives'].values for i in row]

from collections import Counter

narrative_counts = Counter(narratives_list)
tuples_list = [(text, count, count*100/len_dataset) for text, count in narrative_counts.items()]
tuples_list.sort(key=lambda x: x[1], reverse=True)

narratives_list=[]
for idx, row in df.iterrows():
    c=row['Count']
    for narr in row['narratives']:
        narratives_list.append((narr, c))

# Agrupar y sumar
counter = Counter()
for text, value in narratives_list:
    counter[text] += value

# Pasar a lista de tuplas
merged = list(counter.items())
merged.sort(key=lambda x: x[1], reverse=True)
```

```python
import matplotlib.pyplot as plt

# Nmero de elementos a mostrar
n = 20  
top_n = merged[:n]

# Extraer datos
labels = [t[0] for t in top_n]
percentages = [t[1]*100/total_interaction for t in top_n]  # multiplicamos por 100 para mostrar en %

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_n)), percentages)
plt.xticks(range(len(top_n)), labels, rotation=90)  # texto en vertical
plt.ylabel("Percentage (%)")
plt.title(f"Top {n} Narratives by Percentage of Occurrence")
plt.tight_layout()
plt.show()
```

## Objectives

```python
df=pd.read_pickle('/data/jfraile/Programs/COP27/datasets_with_answer/objectives_cop_no_guide_Tweets_per user_id-first_peak_7_11_to_12_11_2022.csv.pkl')
len_dataset=len(df)
df['Count'] = df['Count'].str.replace(',', '').astype(float).astype(int)
total_interaction=df['Count'].sum()
objectives_list=[i for row in df['objectives'].values for i in row]

from collections import Counter

objectives_counts = Counter(objectives_list)
tuples_list = [(text, count, count*100/len_dataset) for text, count in objectives_counts.items()]
tuples_list.sort(key=lambda x: x[1], reverse=True)

objectives_list=[]
for idx, row in df.iterrows():
    c=row['Count']
    for narr in row['objectives']:
        objectives_list.append((narr, c))

# Agrupar y sumar
counter = Counter()
for text, value in objectives_list:
    counter[text] += value

# Pasar a lista de tuplas
merged = list(counter.items())
merged.sort(key=lambda x: x[1], reverse=True)
```

```python
import matplotlib.pyplot as plt

# Nmero de elementos a mostrar
n = 10  
top_n = merged[:n]

# Extraer datos
labels = [t[0] for t in top_n]
percentages = [t[1]*100/total_interaction for t in top_n]  # multiplicamos por 100 para mostrar en %

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_n)), percentages)
plt.xticks(range(len(top_n)), labels, rotation=90)  # texto en vertical
plt.ylabel("Percentage (%)")
plt.yticks(range(0, 55, 5))
for y in range(0, 55, 5):  # from 0% to 100% every 10
    plt.axhline(y=y, color='grey', linestyle='--', linewidth=0.5)
plt.title(f"Top {n} Objectives by Percentage of Occurrence")
plt.tight_layout()
plt.show()
```
